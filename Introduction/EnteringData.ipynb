```r
# Entering data

# Colon Operator

# Assign number 0 through 10 to x1
x1 <- 0:10
x1
```

```r
# Descending Order
x2 <- 10:0
x2
```

```r
# SEQ
?seq # R help on seq
```

```r
# Ascending values (duplicates 1:10)

(x3 <- seq(10))
```

```r
# Specify change in values
(x4 <- seq(30,0, by = -3))
```

```r
# Enter Multiple Values With c

# c = concatenate (or combine or collect)
?c # R help on c
```

```r
x5 <- c(5, 4, 1, 6, 7, 2, 2, 3, 2, 8)
x5
```
