```r
# Principal Component Analysis (PCA)

# Less noise & fewer unhelpful variables in data = more meaning

# Go from 2D to 1D but maintained the most important information

# INSTALL AND LOAD PACKAGES ################################

library(datasets)  # Load base packages manually

# Installs pacman ("package manager") if needed
if (!require("pacman")) install.packages("pacman")

# Use pacman to load add-on packages as desired
pacman::p_load(pacman, tidyverse) 
```

```r
# Load Data
head(mtcars)
```

```r
cars <- mtcars[, c(1:4,6:7,9:11)]  # Select variables
head(cars)
```

```r
# Compute PCA

# For entire data frame
pc <- prcomp(cars,
            center = TRUE, # Center means to 0 (optional)
            scale = TRUE) # Set unit variance (helpful)
pc
```

```r
# OR 
# To specify variables

pc <-prcomp(~ mpg + cyl + disp + hp + wt + qsec + am + gear + carb,
            data = mtcars,
            center = TRUE,
            scale = TRUE)
pc
```

```r
# Examine Results

# Get summary stats
summary(pc)
```

```r
# ScreePlot for number of variables
plot(pc)
```

```r
pc
```

```r
# See how cases load on PCs
predict(pc) %>% round(2)
```

```r
# Biplot of first two components
biplot(pc)
```
