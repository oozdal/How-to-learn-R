# Quantmode package to access Yahoo finance data

```r
# Load quantmode package and plot market data

install.packages("quantmod")
library(quantmod)
library(TTR)
```

```r
getSymbols("AAPL", src="yahoo")
```

```r
chartSeries(AAPL, theme = "black")
```

```r
chartSeries(AAPL, theme = "black")
addEMA(20, col = "red")  # Adding Exponential Moving Average
```

```r
chartSeries(AAPL['2021'], theme = "black")
addEMA(20, col = "red")  # Adding Exponential Moving Average
```

```r
chartSeries(AAPL['2021-04'], theme = "black")
```
